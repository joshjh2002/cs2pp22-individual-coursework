- A **network** or **graph**, $G$, consists of a set of **nodes** (or vertices), $V$, and edges, $E$.
$$$$

- An edge is a pair of nodes $(a,b)$ denoting the nodes connected by the edge.

$$ G = (V,E) $$

![CompareNetworks.png](images/CompareNetworks.png)

- Networks can be represented in various different structures.
$$$$

- We can represent networks using a simple list of all the nodes and all the edges in the network:

$$ V = {a,b,c,d,e}  $$

![Edges.png](images/Edges.png)

- Another data structure that we could use is to build a **neighbour list** for every node.
$$$$

- For every node, $x$, we maintain a list of all the neighbours, $y$.
![Neighbours.png](images/Neighbours.png)

**Data Reference:**

The file `data/dolphins.tsv` contains a representation of a social network dataset where dolphins have links between them if they frequently associated with one another.

To read in the *tab separated* network file, we need to read each line in the file. To do this we just treat the file object as an iterator.
- This is an example of using a context manager: https://book.pythontips.com/en/latest/context_managers.html
$$$$

- Here `line` will be a string for each line in the file. The `.split(x)` method splits a string into a list of substrings for each occurrence of the character `x` (in this case the tab: `\t`).


This code reads the file and places the data in an edge list representation:
```python
# Create an empty edge set
edges = set()

    for line in file:
        a, b = line.split('\t')
        e = (int(a), int(b))
        edges.add(e)
```

    - There should be **no network data output from this cell**.$$$$

There will be many nodes in the final dataset.  The result should have the form:
```python
{55: {2, 7, 8, 14, 20, 42, 58},
 2: {18, 20, 27, 28, 29, 37, 42, 55},
 20: {2, 8, 31, 55},...
 
```

This shows that dolphin `55` is frequently associated with dolphins `2, 7, 8, 14, 20, 42, and 58`.

````python
# Begin with edges

edges = set()

with open('data/dolphins.tsv', 'r') as file:
    for line in file:
        a, b = line.split('\t')
        e = (int(a), int(b))
        edges.add(e)

dict = {}

for edge in edges:
    if edge[0] not in dict:
        dict[edge[0]] = set() # If the first edge is not in the dictionary, initialise the key-value pair
    dict[edge[0]].add(edge[1]) # Add the edge to the dictionary
    if edge[1] not in dict:
        dict[edge[1]] = set()
    dict[edge[1]].add(edge[0])
````

````python
# Verifying my more efficient solution

dict
````

````python
%%timeit -r 19 -n 2000
# Testing the performance of my more efficient solution

dict = {}

for edge in edges:
    if edge[0] not in dict:
        dict[edge[0]] = set()
    dict[edge[0]].add(edge[1])
    if edge[1] not in dict:
        dict[edge[1]] = set()
    dict[edge[1]].add(edge[0])

# Testing time to get the neighbors of a node too
dict[55]
````

**Discussion:**



My solution is to convert the set of edges into a dictionary using a for loop. Each itteration checks if the dictionary already contains either part of the current edge set tuple and adds them to the dictionary. It then appends both the values in the tuple to dictionary where they key is the other value in the tuple. 

The average time for this to complete is ~40µs. This method has an O(n) time complexity as it requires a for loop to iterate through the set of edges. This means that the time taken to run the code will increase linearly with the size of the dataset. As such, the larger the dataset, the longer the code will take to run.

````python
# Verifying my less efficient solution

# print(max(edges)) # largest node is 62

# This function takes in a matrix and a node
def get_connections(mat, node):
    # Empty list
    connections = []
    # For all nodes in the 1d array in mat[node-1]
    for i in range(len(mat[node-1])):
        # 1 means it's connected
        if mat[node-1][i] == 1:
            # Append i + 1 because i is always 1 less than the node
            connections.append(i+1)
    return connections

# Initialise 2D matrix where all values are set to 0
adjacency_matrix = [[0 for i in range(62)] for j in range(62)]

# Iterates over all edges and sets the connection to 1
for edge in edges:
    adjacency_matrix[edge[0]-1][edge[1]-1] = 1
    adjacency_matrix[edge[1]-1][edge[0]-1] = 1

# list all connections to node 55
print(get_connections(adjacency_matrix, 55), get_connections (adjacency_matrix, 2), get_connections(adjacency_matrix, 20))
````

````python
%%timeit -r 19 -n 2000
# Testing the performance of my less efficient solution -- testing setting up the adjacency matrix and getting connections

adjacency_matrix = [[0 for i in range(62)] for j in range(62)]

for edge in edges:
    adjacency_matrix[edge[0]-1][edge[1]-1] = 1
    adjacency_matrix[edge[1]-1][edge[0]-1] = 1

get_connections(adjacency_matrix, 55)
````

**Discussion:**



This method uses an adjacency matrix. It creates a matrix of zeros with the size of the matrix being the number of nodes in the dataset. It then iterates through the set of edges and changes the value of the matrix at the index of the two nodes in the edge set to 1. To look up the neighbours of a node, it iterates through the row of the matrix at the index of the node and adds the index of any column that has a value of 1. This takes around 100 µs to complete the matrix setup and lookup the connections to node 55. This is almost 60 µs longer than the dictionary method.

#### 2.2.1. Add `HP_Type` [2 marks]

Using an implementation of **list comprehension**, create a new column, `HP_Type`, such that,
- if a car's `HP` is greater than or equal to 300, `HP_Type` is 'high'.


- Otherwise, `HP_Type` is 'low'.

#### 2.2.2.  Add `Price_class` [2 marks]

Using an implementation of a **function**, create a new column, `Price_class`, such that it becomes equal to:
- 'high', if `Price` is greater than or equal to 50,000
- 'mid', if `Price` is between 30,000 (inclusive) and 50,000 (exclusive), and
- 'low', if `Price` is below 30,000.
    
Then, display the resulting `DataFrame`.

````python
def calculate_price_class(price):
    if price >= 50000:
        return 'high'
    elif price >= 30000:
        return 'mid'
    else:
        return 'low'


cars['Price_class'] = cars['Price'].apply(calculate_price_class)

cars.head()
````

Design a `Tournament` class to represent **teams** of car efficiency enthusiasts that can collect an **inventory** of fuel efficient cars by finding sponsorship from car makers. 

Once **sponsors** are arranged and teams are formed, teams will **compete** by facing off head-to-head (pairwise, bracket-style) until only one `Tournament.champion` prevails.

#### 3.2. Define a `Tournament` class [30 marks]


Your are to translate **these requirements** (with consideration of their usage, defined in **3.3** and **3.4**) into working code.



**Overall**: Coding efficiency and structure, including comments and docstrings, where appropriate, will contribute to the mark in this section.

- The class is initialised with:
    - a `DataFrame` of **cars**, as designed above
        - **No missing values** are permitted.
    - a tournament **name**
    - an **optional number of competing teams**, defaulting to 16.
        - If the input number of teams is not an integer, `raise` the appropriate kind of exception, and a message saying, "The number of teams must be an integer."
        - Also, `assert` that the value is positive and non-zero.
        - Ensure that this number is a **perfect square**.
    - Include sensible object representation dunder methods (i.e. `__repr__` and `__str__`).
$$$$
- There is a method to `generate_sponsors`.
    - The method, by default, **randomly selects a sponsor** for each team from the available list of `Make`rs (with no duplicates).
    - It **accepts a list of specified sponsors as an optional argument** with a length up to the number of teams.
        - If the supplied list is shorter, the remainder of the teams will be set random sponsors, as in the default case.
    - Once sponsors are selected, the sponsor **randomly allocates a budget** in some suitably chosen range of incremental integer values defined by `low`, `high`, and `incr` parameters to this `generate_sponsors` method.  Choose sensible default values for these.
    - Store the resulting sponsor and budget information as attributes of the `Tournament` class.$$$$    

- There is a method to `generate_teams`.
    - This method should simply populate a list of `Team`s in the `Tournament` class.$$$$
    

- The teams are members of the `Team` class, a class **internal to the `Tournament` class**.
    - A `Team` object holds information about its:
        - sponsor
        - budget
        - inventory
        - `active` status, i.e. whether they are eligible to compete in the next round of the tournament.
        - performance record (e.g., win/loss, score, number of cars they used in competition)
    - This class has some kind of `__str__` representation.
    - Be thoughtful about which of these characteristics are mutable and how to change them if they are not mutable.
$$$$
- There is a method to `buy_cars`.
    - This method will allow the `Team`s to each purchase their initial inventory.
$$$$
- There is a method to `_purchase_inventory`.
    - This internal method takes 1 argument: a single `Team` object.
    - With the information provided by the `Team`, isolate cars that are available from the **sponsoring `Make`r**.  
    - Considering the `Team` budget, select the set of cars representing the **optimal choice**.  That is, purchase as many cars as the budget permits, while trying to **maximise `MPG-H`**.  
        - For more comprehensive implementations of this combinatorial optimisation problem, expect this to be the **most computationally expensive** part of this sub-task.  It can reasonably require a few seconds to perform the optimal selection.
    - When these are selected, the `Team`'s budget and inventory should be updated.
$$$$    
- There is a method to `hold_event` (i.e., execute the tournament competition process).
    - Cycle through the pairwise matches, keeping track of `Team` performance metrics.
        - The teams will compete in a head to head match and either continue on or be **eliminated**.
    - After each match, allocate a **financial prize** to the winning `Team`.  You can decide how to implement this; perhaps the prize increases in every round, but it must be enough to allow them to take the next step...
    - After awarding the prize, allow the `Team` to `_purchase_inventory` again (increasing the number of cars in their inventory) before the next match.  
        - Newly purchased cars can be duplicates of members of the `Team`'s existing inventory, but only one any kind of car is permitted to be purchased in an individual `_purchase_inventory` action.
    - At the end of the tournament event, record the `Tournament.champion` `Team`.  
    

````python
# Imports required libraries. Warnings is to suppress warnings about future changes to pandas
import random
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)


class Team: 
    """
    Team object which contains all the information about a team.

    Attributes:
    - sponsor (str): The sponsor of the team.
    - budget (int): The budget of the team.
    - inventory (pd.DataFrame): The DataFrame containing the cars the team has purchased.
    - active (bool): Whether the team is still active in the tournament.
    - performance_record (list): The list of the team's performance in the tournament.
    - current_max_distance (int): The current maximum distance the team has achieved.
    """
    
    def __init__(self, sponsor: str, budget: int):
        self.sponsor = sponsor
        self.budget = budget
        self.inventory = pd.DataFrame()
        self.active = True
        self.performance_record = list()
        self.current_max_distance = 0

    def __str__(self):
        return f'is the {self.sponsor} team with a budget of ${self.budget} and {self.inventory.shape[0]} cars'
    
    def evaluate_performance(self):
        if len(self.inventory) == 0:
            self.active = False
            return
        self.current_max_distance = self.inventory['MPG-H'].sum()


class Tournament:
    """
    Represents a tournament where teams compete against each other.
    
    Attributes:
    - cars (pd.DataFrame): The DataFrame containing car information.
    - name (str): The name of the tournament.
    - number_of_teams (int): The number of teams participating in the tournament.
    - sponsors (list): The list of sponsors for the teams.
    - budgets (list): The list of budgets for the teams.
    - teams (list): The list of teams participating in the tournament.
    - records (list): The list of the teams who have lost.
    - champion (Team): The champion team of the tournament.
    - prize (int): The prize amount for winning the match.
    """
    
    def __init__(self, cars: pd.DataFrame, name: str, number_of_teams=16):
        if cars.isnull().any().any():
            raise ValueError('DataFrame cannot contain null values')
        
        if type(number_of_teams) != int:
            raise TypeError('Team size must be an integer')

        if number_of_teams <= 0:
            raise ValueError('Team size must be greater than 0')

        if number_of_teams ** 0.5 % 1 != 0:
            raise ValueError('Team size must be a perfect square')

        self.cars = cars
        self.name = name

        self.number_of_teams = number_of_teams

        self.__sponsors = []
        self.__budgets = []
        self.__teams = list()
        self.__records = list()
        self.champion = None

        # The prize starts at 10,000 and increases by 10,000 after each match
        self.__prize = 10_000

    def __repr__(self):
        return f'{type(self).__qualname__}({self.cars}, {self.name}, {self.number_of_teams}, {self.__sponsors}, {self.__budgets})'

    def __str__(self):
        return f'Tournament: {self.name} | Team Count: {self.number_of_teams}\nSponsors: {self.__sponsors}\nBudgets: {self.__budgets}'

    def generate_sponsors(self, sponsors: list = list(), low=200_000, high=1_000_000, increment=5_000):
        if len(sponsors) > self.number_of_teams:
            raise ValueError(
                'Number of sponsors cannot be greater than the team count')
        
        # Generate a list of unique sponsors
        potential_sponsors = list(self.cars['Make'].unique())

        # Takes the already passed sponsors and adds more sponsors until it reaches the team count
        while len(sponsors) < self.number_of_teams:
            random_sponsor = random.choice(potential_sponsors)
            sponsors.append(random_sponsor)
            potential_sponsors.remove(random_sponsor)

        self.__sponsors = sponsors

        budgets = []

        # Generates a random budget for each sponsor
        for sponsor in sponsors:
            budgets.append(random.randrange(low, high, increment))

        self.__budgets = budgets

    def generate_teams(self):
        teams = list()

        for sponsor, budget in zip(self.__sponsors, self.__budgets):
            teams.append(Team(sponsor, budget))

        self.__teams = teams

    def buy_cars(self):
        for team in self.__teams:
            self.__purchase_inventory(team)

    ## This uses a greedy approach by the car with the highest MPG-H first
    def __purchase_inventory(self, team: Team):
        cars = self.cars[self.cars['Make'] == team.sponsor].sort_values(axis=0, by='MPG-H', ascending=False)

        # Iterates over the cars and purchases them until the team runs out of money or cars
        while team.budget > 0 and cars.__len__() > 0:
            # Gets the first car in the list and its price (the highest MPG-H)
            car = cars.iloc[0]
            car_price = car['Price']

            # Checks if the team can afford the car
            if team.budget - car_price >= 0:
                # If they can, the car is added to the team's inventory and the budget is updated
                team.inventory = team.inventory.append(car)
                team.budget -= car_price

                # The car is removed from the list of cars as not to purchase the same car twice
                cars = cars.iloc[1:]
            else:
                break
        
        team.evaluate_performance()

    def hold_event(self):

        # Creates an empty list of matches
        matches = list()

        while len(self.__teams) > 1:
            # Creates a list of tuples of teams to match up
            for i in range(0, len(self.__teams), 2):
                matches.append((self.__teams[i], self.__teams[i+1]))

            # Iterates over the list of matched teams and holds a match between them     
            while len(matches) > 0:
                team1, team2 = matches.pop()

                winner, loser = self.__hold_match(team1, team2)

                # The loser is removed from the teams list and added to the records list
                self.__records.append(loser)
                self.__teams.remove(loser)

                # The winner is given the prize money and the inventory is updated
                winner.budget += self.__prize
                self.__purchase_inventory(winner)

            # The prize money increases by 5,000 after each match
            self.__prize += 10_000
        
        # The last team standing is the champion
        self.champion = self.__teams[0]
        self.__records.append(self.champion)

    def __hold_match(self, team1: Team, team2: Team):
        # Compares the current max distance of each team and returns the winner and loser. Also appends to their performance records and deactivates the loser
        if team1.current_max_distance > team2.current_max_distance:
            team1.performance_record.append('W')
            team2.performance_record.append('L')
            team2.active = False
            return team1, team2
        elif team1.current_max_distance < team2.current_max_distance:
            team1.performance_record.append('L')
            team1.active = False
            team2.performance_record.append('W')
            return team2, team1
    
    def show_win_records(self):
        for team in self.__records:
            print(f'{team.sponsor}: {team.performance_record}')
        
        print(f'{self.champion.sponsor} is the champion!')
````
